```python
import numpy as np
import pickle
import tensorflow as tf
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences]) # fetching maximum length sentence
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=279)
y.shape
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense, Bidirectional
```

```python
model_lstm = Sequential()
model_lstm.add(Embedding(279, 100, input_length=194))
model_lstm.add(LSTM(50))
model_lstm.add(Dense(279, activation='softmax'))

model_lstm.summary()
```

```python
model_lstm.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model_lstm.fit(X,y,epochs=50)
```

```python
model_lstm.fit(X,y, initial_epoch=50 , epochs=100)
```

# Using stacked LSTM

```python
model_st_lstm = Sequential([
    Embedding(283, 100, input_length=56),
    LSTM(150, return_sequences=True),
    LSTM(150, return_sequences=True),
    LSTM(150, return_sequences=True),
    LSTM(150),
    Dense(283, activation='softmax')
])

model_st_lstm.summary()

model_st_lstm.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model_st_lstm.fit(X,y,epochs=100)
```

# Model using stacked GRU

```python
# Define the GRU model_st_GRU
model_st_GRU = Sequential([
    Embedding(283, 100, input_length=56),
    GRU(100),
    Dense(283, activation='softmax')
])

model_st_GRU.summary()

model_st_GRU.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model_st_GRU.fit(X,y,epochs=100)
```

```python
# Define the GRU model_st_GRU
model_st_GRU = Sequential([
    Embedding(283, 100, input_length=56),
    GRU(10, return_sequences=True),
    GRU(10, return_sequences=True),
    GRU(10, return_sequences=True),
    GRU(10),
    Dense(283, activation='softmax')
])

model_st_GRU.summary()

model_st_GRU.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model_st_GRU.fit(X,y,epochs=100)
```
